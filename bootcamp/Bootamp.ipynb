# A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary. It does not show a distribution in as much detail as a stem and leaf plot or histogram does, but is especially useful for indicating whether a distribution is skewed and whether there are potential unusual observations (outliers) in the data set. 

# A boxplot is plotted  by invoking .boxplot() on the DataFrame.

# DGS Box plots:


# Using the boxplot graph, we can compare the range and distribution of the  DGS and other variables . We observe that there is a greater variability for DGS as well as larger outliers.

# Also, since the notches in the boxplots do not overlap, we can conclude that with 95% confidence, that the true medians do differ.

# Most outliers of the DGS Box plots are below 2.5
